```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
import floris.tools as wfct
```

# Floris-modeled wake

```python
# Initialize the FLORIS interface fi
fi = wfct.floris_utilities.FlorisInterface("V27.json")
```

```python
# Calculate wake
fi.calculate_wake()
```

```python
turbine = fi.floris.farm.turbines[0]
```

```python
flowdata = fi.get_flow_data(grid_spacing=1.25)
```

```python
fig, ax = plt.subplots()
hor_plane = wfct.cut_plane.HorPlane(flowdata, turbine.hub_height)
wfct.visualization.visualize_cut_plane(hor_plane, ax=ax)
```

```python
for downD in range(2,9):
    fig, ax = plt.subplots()
    cross_plane = wfct.cut_plane.CrossPlane(flowdata, downD*turbine.rotor_diameter)
    wfct.visualization.visualize_cut_plane(cross_plane, ax=ax)
```
