### define filepath

```python
filepath = "/Volumes/pdoubraw-1/repos/wakebench/data/neutral/lidar.z01.b0.20170123.225000.mat"
stability= "neutral"
key      = "20170123_225000"
```

### import packages

```python
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
fs=16
plt.rc("font",size=fs)
```

```python
# make sure you clone https://github.com/ewquon/waketracking.git and add it to your python path
import sys
sys.path.append("/Volumes/pdoubraw-1/repos/waketracking")
from samwich.waketrackers import track
from samwich.dataloaders import SpinnerLidarMatlab
```

### some fixed parameters

```python
# rotor diameter of V27
D     = 27.0
# hub height of V27
z_hub = 32.1
# rotor radius
R      = D/2.
# reference area for tracking the wake
ref_area = np.pi/4 * D**2
# parameters to reconstruct 10-min mean vertical wind speed profile via a power law 
# these were derived from sonic measurements at the mast
power_law_params = {'neutral':{
                    '20170123_225000' : {'uref': 8.7, 'zref':58.0, 'alpha':0.13},
                    '20170622_012000' : {'uref': 9.2, 'zref':58.0, 'alpha':0.11},
                    '20170622_013000' : {'uref': 9.2, 'zref':58.0, 'alpha':0.13},
                    '20170622_014000' : {'uref': 9.4, 'zref':58.0, 'alpha':0.17},
                    '20170622_015000' : {'uref': 9.7, 'zref':58.0, 'alpha':0.17},
                    '20170703_141000' : {'uref': 9.8, 'zref':58.0, 'alpha':0.13}
                              },
                    'unstable':{
                        '20170217_144000':{'uref': 7.9, 'zref':58.0, 'alpha':0.20},
                        '20170217_145000':{'uref': 7.1, 'zref':58.0, 'alpha':0.14},
                        '20170217_150000':{'uref': 7.2, 'zref':58.0, 'alpha':0.13},
                        '20170217_151000':{'uref': 7.6, 'zref':58.0, 'alpha':0.14},
                        '20170217_153000':{'uref': 7.9, 'zref':58.0, 'alpha':0.09}
                               },
                    'stable':{
                        '20170706_072000':{'uref': 5.5, 'zref':58.0, 'alpha':0.44},
                        '20170706_073000':{'uref': 5.6, 'zref':58.0, 'alpha':0.49},
                        '20170706_074000':{'uref': 5.5, 'zref':58.0, 'alpha':0.56},
                        '20170708_081000':{'uref': 7.0, 'zref':58.0, 'alpha':0.49},
                        '20170708_082000':{'uref': 7.4, 'zref':58.0, 'alpha':0.48},
                        '20170708_083000':{'uref': 8.1, 'zref':58.0, 'alpha':0.56}
                             }
                   }
```

### read the matlab file

```python
lidar_data_all = SpinnerLidarMatlab(filepath, verbose=True)
```

### for each downstream distance scanned, interpolate the scan points into a regular grid

```python
available_x      = np.unique(lidar_data_all.scan_avg.focus_dist_set_D)
# only do it if x/D > 1.5, otherwise the scan doesn't usually cover the entire wake anyway
available_x      = available_x[available_x>1.5]
# fow now, skip the distance x/D = 2.5 which is only available in some cases and was not provided by modelers
available_x      = available_x[available_x!=2.5]    
# a new lidar object needs to be created for each distance we want to interpolate (this is just the way the code was set up)        
# allocate space in a dictionary to keep these data objects separate but organized
lidar_data_per_x = {}
for x in available_x:
    print("x/D ~ {0:.0f}".format(x))
    lidar_data_per_x[x] = SpinnerLidarMatlab(filepath, verbose=True)
    lidar_data_per_x[x].interpolate(focaldist_D=x, force2D=True, ds=1.0, horzrange=[-60,60], vertrange=[0,120])
```

### perform wake tracking

```python
# define a y/z vector for regridding data into a meandering frame of reference within the algorith,
y_mfor = np.arange(-50,50.1,1)
z_mfor = np.arange(-40,40.1,1)
# allocate space to keep the wake objects generated by the algorithm -- one object per downstream distance 
wake   = {}
for x in lidar_data_per_x.keys():
    wake[x] = track(lidar_data_per_x[x].sliceI(), method='Gaussian2D', verbose=0)
    wake[x].remove_shear(alpha=power_law_params[stability][key]['alpha'],Uref=power_law_params[stability][key]['uref'],zref=power_law_params[stability][key]['zref'])
    ywake, zwake = wake[x].find_centers(A_ref=ref_area,
                                        A_min=0.5*ref_area,
                                        A_max=2.0*ref_area,
                                        AR_max=3)
    wake[x].to_MFoR(y_mfor,z_mfor)        
```

### visualize mean

```python
def plot_contours(wake_dictionary, component_name, vmin, vmax, label):
 
    distances = wake_dictionary.keys()
 
    fig = plt.figure(figsize=(len(distances)*5+2,5))
    
    axs = {}
    
    for iax,distance in enumerate(distances):

        axs[iax]  = fig.add_subplot(1,len(distances),iax+1,aspect='equal')

        if "mfor" in component_name:
            xhnorm = (wake_dictionary[distance].xh_mfor)/D
            xvnorm = (wake_dictionary[distance].xv_mfor)/D
        else:
            xhnorm    = wake_dictionary[distance].xh/D
            xvnorm    = (wake_dictionary[distance].xv-z_hub)/D
        uu        = np.nanmean(getattr(wake_dictionary[distance],component_name),axis=0)

        p = axs[iax].pcolormesh(xhnorm,
                            xvnorm,
                            uu,
                            vmin=vmin, vmax=vmax)
        axs[iax].invert_xaxis()
        axs[iax].set_xlim([-1.3,1.3])
        axs[iax].set_xticks([1,0,-1])
        axs[iax].set_ylim([-1.3,1.3])
        axs[iax].set_yticks([-1,0,1])
        axs[iax].grid(color='gray',linestyle=':')
        axs[iax].xaxis.set_tick_params(direction='in')
        axs[iax].yaxis.set_tick_params(direction='in')       

        ylabel = "$y_{mfor}/D$ [-]" if "mfor" in component_name else "$y_{ffor}/D$ [-]"
        zlabel = "$z_{mfor}/D$ [-]" if "mfor" in component_name else "$z_{ffor}/D$ [-]"        
        
        axs[iax].set_xlabel(ylabel)
        if iax>0:
            plt.setp(axs[iax].get_yticklabels(), visible=False)
        else:
            axs[iax].set_ylabel(zlabel)    

        axs[iax].set_title("$x/D\sim${0}".format(distance),fontsize=fs)
  
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.82, 0.2, 0.01, 0.6])
    clb = fig.colorbar(p, cax=cbar_ax, format=FormatStrFormatter('%4.1f'))
    clb.set_label(label, labelpad=15)
    clb.ax.tick_params(direction='in')     
    fig.subplots_adjust(wspace=0.1)

    return fig
```

```python
fig = plot_contours(wake, component_name="u_tot", vmin = 1.0, vmax = 9.0, label=r'Projected line of sight [m/s]')
```

```python
fig = plot_contours(wake, component_name="u", vmin = -4.0, vmax = 2.0, label=r'Velocity deficit [m/s]')
```

```python
fig = plot_contours(wake, component_name="u_mfor", vmin = -4.0, vmax = 2.0, label=r'Velocity deficit [m/s]')
```

### visualize snapshots

```python
def plot_contours_snapshot(wake_dictionary, time_index, component_name, vmin, vmax, label):
 
    distances = wake_dictionary.keys()
 
    fig = plt.figure(figsize=(len(distances)*5+2,5))
    
    axs = {}
    
    for iax,distance in enumerate(distances):

        axs[iax]  = fig.add_subplot(1,len(distances),iax+1,aspect='equal')

        if "mfor" in component_name:
            xhnorm = (wake_dictionary[distance].xh_mfor)/D
            xvnorm = (wake_dictionary[distance].xv_mfor)/D
        else:
            xhnorm    = wake_dictionary[distance].xh/D
            xvnorm    = (wake_dictionary[distance].xv-z_hub)/D
        
        uu        = getattr(wake_dictionary[distance],component_name)
        uu        = uu[time_index,...].copy()

        if "mfor" in component_name:
            wake_y = 0.0
            wake_z = 0.0
        else:
            wake_y = wake_dictionary[distance].xh_wake[time_index]/D
            wake_z = (wake_dictionary[distance].xv_wake[time_index] - z_hub)/D
        axs[iax].plot(wake_y, wake_z, '+w', ms=8, mew=2)
        
        if "mfor" in component_name:
            wake_edge_y = wake_dictionary[distance].paths_mfor[time_index][:,0]/D
            wake_edge_z = wake_dictionary[distance].paths_mfor[time_index][:,1]/D      
        else:
            wake_edge_y = wake_dictionary[distance].paths[time_index][:,0]/D
            wake_edge_z = (wake_dictionary[distance].paths[time_index][:,1]-z_hub)/D        
                    
        axs[iax].plot(wake_edge_y, wake_edge_z, '-w', lw=2)
        
        p = axs[iax].pcolormesh(xhnorm,
                            xvnorm,
                            uu,
                            vmin=vmin, vmax=vmax)
        
        axs[iax].invert_xaxis()
        axs[iax].set_xlim([-1.3,1.3])
        axs[iax].set_xticks([1,0,-1])
        axs[iax].set_ylim([-1.3,1.3])
        axs[iax].set_yticks([-1,0,1])
        axs[iax].grid(color='gray',linestyle=':')
        axs[iax].xaxis.set_tick_params(direction='in')
        axs[iax].yaxis.set_tick_params(direction='in')       

        ylabel = "$y_{mfor}/D$ [-]" if "mfor" in component_name else "$y_{ffor}/D$ [-]"
        zlabel = "$z_{mfor}/D$ [-]" if "mfor" in component_name else "$z_{ffor}/D$ [-]"        
        
        axs[iax].set_xlabel(ylabel)
        if iax>0:
            plt.setp(axs[iax].get_yticklabels(), visible=False)
        else:
            axs[iax].set_ylabel(zlabel)    

        axs[iax].set_title("$x/D\sim${0}".format(distance),fontsize=fs)
  
    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.82, 0.2, 0.01, 0.6])
    clb = fig.colorbar(p, cax=cbar_ax, format=FormatStrFormatter('%4.1f'))
    clb.set_label(label, labelpad=15)
    clb.ax.tick_params(direction='in')     
    fig.subplots_adjust(wspace=0.1)

    return fig
```

```python
fig = plot_contours_snapshot(wake, time_index=5, component_name="u_tot", vmin = 1.0, vmax = 9.0, label=r'Projected line of sight [m/s]')
```

```python
fig = plot_contours_snapshot(wake, time_index=5, component_name="u", vmin = -4.0, vmax = 2.0, label=r'Velocity deficit [m/s]')
```

```python
fig = plot_contours_snapshot(wake, time_index=5, component_name="u_mfor", vmin = -4.0, vmax = 2.0, label=r'Velocity deficit [m/s]')
```
